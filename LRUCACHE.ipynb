```python
## LRU CACHE DEMO
```

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=128)
def fibb(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    return fibb(n-2)+fibb(n-1)
```

```python
%timeit fibb(10)
```

```python
%timeit fibb(100)
```

```python
%timeit fibb(1000)
```
