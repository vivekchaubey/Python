```python
def null_decorator(func):
    return func
```

```python
def hello():
    return "Hello"
```

```python
hello()
```

```python
dec_hello=null_decorator(hello)
```

```python
dec_hello()
```

```python
## COMPLICATED DECORATOR
## LETS ADD WRAPPER TO A STRING
```

```python
def emphasise(func):
    def wrapper(*args,**kws):
        string=func(*args,**kws)
        output="<h>"+string+"</h>"
        return output
    return wrapper
```

```python
## @emphasis works as a wrapper for the function
@emphasise
def test():
    return "test"
```

```python
test()
```

```python
### LETS DEMO OPTIONAL ARGUMENTS IN WRAPPERS
```

```python
def multiling(lang="en"):
    lookup={"en":"hello","fr":"bonjour"}
    return lookup[lang]
```

```python
multiling("en")
```

```python
multiling("fr")
```

```python
multiling()
```

```python
@emphasise
def multiling2(lang="en"):
    lookup={"en":"hello","fr":"bonjour"}
    return lookup[lang]
```

```python
multiling2()
```

```python
multiling2("fr")
```
