```python
def fibb(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    return fibb(n-1)+fibb(n-2)
```

```python
%timeit fibb(12)
```

```python
%timeit fibb(32)
```

```python
%timeit fibb(32)
```

```python
def memoize(func):
    cache=dict()
    def memoize_func(arg):
        if arg in cache:
            return cache[arg]
        else:
            result=func(arg)
            cache[arg]=result
            return result
    return memoize_func
```

```python
mfib=memoize(fibb)
```

```python
%timeit mfib(10)
```

```python
%timeit mfib(25)
```

```python
%timeit mfib(25)
```
